```python
data_location = pd.read_csv('/content/gdrive/MyDrive/location.csv')
df_location = pd.DataFrame(data_location)
df_location['갯수'] = list_count
df_location.head()
```

```python
# import the library
import folium

# 서울시 중심부의 위도, 경도 입니다.
center = [37.541, 126.986]

# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다.
m = folium.Map(location=center, zoom_start=10)

for i in range(0,len(df_location)):
   folium.Circle(
      location=[df_location.iloc[i]['위도'], df_location.iloc[i]['경도']],
      popup=df_location.iloc[i]['위치'],
      radius=float(df_location.iloc[i]['갯수']) * 10,
      color='crimson',
      fill=True,
      fill_color='crimson'
   ).add_to(m)

m
```
